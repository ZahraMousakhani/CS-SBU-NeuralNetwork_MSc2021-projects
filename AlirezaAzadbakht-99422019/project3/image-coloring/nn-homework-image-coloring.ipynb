# Loading data

```python
IMG_HEIGHT = 256
IMG_WIDTH = 256
TRAIN_PATH = '../input/art-images-drawings-painting-sculpture-engraving/dataset/dataset_updated/training_set/painting/'
train_ids = next(os.walk(TRAIN_PATH))[2]
```

```python
x = []
y = []
for img_id in  train_ids:
    try:
        img = imread(TRAIN_PATH+img_id)
        img = resize(img, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True)
        img = img.astype('float32') / 255.
    except(Exception):
#         print(img_id, 'lost!')
        continue
        
    gray_img = rgb2gray(img).reshape(256,256,1)
    img = np.transpose(img, (2,0,1))
    gray_img = np.transpose(gray_img, (2,0,1))
    x.append(gray_img)
    y.append(img)

print('x shape',len(x))
print('y shape',len(y))
```

```python
x_train, x_test, y_train, y_test = train_test_split(np.array(x), np.array(y), test_size=0.10, random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
class MyDataset(torch.utils.data.Dataset):
    def __init__(self, x, y):
        self.x = torch.tensor(x)
        self.y = torch.tensor(y)

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        x = self.x[idx]
        y = self.y[idx]
        return (x, y)
    
BATCH_SIZE = 32    
train_dataset = MyDataset(x_train, y_train)
dataLoader_train = torch.utils.data.DataLoader(train_dataset,
                                               batch_size=BATCH_SIZE,
                                               shuffle=True)

test_dataset = MyDataset(x_test, y_test)
dataLoader_test = torch.utils.data.DataLoader(test_dataset,
                                              batch_size=BATCH_SIZE,
                                              shuffle=False)
```

# Modeling

```python
def make_callback(s_key, model_name):
    temp = imaging.MakeGrid(s_key, num_images=10, nrow=10)
    temp = temp.on_test().to_pyplot().to_file('{}_{}.png'.format(model_name,str(s_key)))
    return temp
```

## shallow

```python
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()

        self.encoder = nn.Sequential(
            nn.Conv2d(1, 32, 4, 1, 2),   # B,  32, 28, 28
            nn.ReLU(True),
            nn.Conv2d(32, 32, 4, 2, 1),  # B,  32, 14, 14
            nn.ReLU(True),
            nn.Conv2d(32, 64, 4, 2, 1),  # B,  64,  7, 7
        )
        
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 4, 2, 1), # B,  64,  14,  14
            nn.ReLU(True),
            nn.ConvTranspose2d(32, 32, 4, 2, 1, 1), # B,  32, 28, 28
            nn.ReLU(True),
            nn.ConvTranspose2d(32, 3, 4, 1, 2)   # B, 1, 28, 28
        )


    def forward(self, x):
        x = self.encoder(x).relu()
        x = self.decoder(x).relu()
        return x
```

### Diffrent Learning rate

```python
model_name = 'shallow_model_mse_0.0001'
model = Autoencoder()

device = 'cuda' if torch.cuda.is_available() else 'cpu'
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.0001)
loss = nn.MSELoss()

trial = Trial(model, optimizer, loss, metrics=['acc', 'loss'],
              callbacks=[callbacks.ConsolePrinter(),
                         make_callback(torchbearer.TARGET, model_name),
                         make_callback(torchbearer.INPUT, model_name),
                         make_callback(torchbearer.PREDICTION, model_name),
                        ]).to(device)

trial.with_generators(train_generator=dataLoader_train, test_generator=dataLoader_test)
history = trial.run(epochs=5, verbose=1)
trial.evaluate(data_key=torchbearer.TEST_DATA)
torch.save(model.state_dict(), './save_{}'.format(model_name))
pd.DataFrame.from_records(history)['loss'].plot()
plt.show()
```

## skip Connection

```python
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()

        self.encoder_l1 = nn.Sequential(
            nn.Conv2d(1, 32, 4, 1, 2),   # B,  32, 28, 28
            nn.ReLU(True)
        )
        self.encoder_l2 = nn.Sequential(
            nn.Conv2d(32, 32, 4, 2, 1),  # B,  32, 14, 14
            nn.ReLU(True),
        )
        self.encoder_l3 = nn.Sequential(
            nn.Conv2d(32, 64, 4, 2, 1),  # B,  64,  7, 7
            nn.ReLU(True)
        )
        
        self.decoder_l1 = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 4, 2, 1), # B,  64,  14,  14
            nn.ReLU(True)
        )
        self.decoder_l2 = nn.Sequential(
            nn.ConvTranspose2d(32, 32, 4, 2, 1, 1), # B,  32, 28, 28
            nn.ReLU(True)
        )
        self.decoder_l3 = nn.Sequential(
            nn.ConvTranspose2d(32, 3, 4, 1, 2),   # B, 1, 28, 28
            nn.ReLU(True)
        )
        self.skip = nn.Identity()


    def forward(self, x):
        x_e1 = self.encoder_l1(x)
        x_e2 = self.encoder_l2(x_e1)
        x_e3 = self.encoder_l3(x_e2)
        x_d1 = self.decoder_l1(x_e3)
        x_d1 = x_d1 + self.skip(x_e2)
        x_d2 = self.decoder_l2(x_d1)
        x_d2 = x_d2 + self.skip(x_e1)
        x_d3 = self.decoder_l3(x_d2)
#         print(x_e1.shape,x_e2.shape,x_e3.shape)
#         print(x_d1.shape,x_d2.shape,x_d3.shape)
        return x_d3
```

```python
model_name = 'U-net_model_mse_0.0001'
model = Autoencoder()

device = 'cuda' if torch.cuda.is_available() else 'cpu'
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.0001)
loss = nn.MSELoss()

trial = Trial(model, optimizer, loss, metrics=['acc', 'loss'],
              callbacks=[callbacks.ConsolePrinter(),
                         make_callback(torchbearer.TARGET, model_name),
                         make_callback(torchbearer.INPUT, model_name),
                         make_callback(torchbearer.PREDICTION, model_name),
                        ]).to(device)

trial.with_generators(train_generator=dataLoader_train, test_generator=dataLoader_test)
history = trial.run(epochs=5, verbose=1)
trial.evaluate(data_key=torchbearer.TEST_DATA)
torch.save(model.state_dict(), './save_{}'.format(model_name))
pd.DataFrame.from_records(history)['loss'].plot()
plt.show()
```

# Salt and Pepper

```python
def sp_noise(image,prob, thres):
    output = np.zeros(image.shape)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn < thres:
                output[i][j] = 1
            else:
                output[i][j] = image[i][j]
    return output
```

```python
x = []
y = []
for img_id in  train_ids:
    try:
        img = imread(TRAIN_PATH+img_id)
        img = resize(img, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True)
        img = img.astype('float32') / 255.
    except(Exception):
#         print(img_id, 'lost!')
        continue
    rdn = random.random()/4
    salt_pepper_img = sp_noise(img, rdn, rdn*2)
    img = np.transpose(img, (2,0,1))
    salt_pepper_img = np.transpose(salt_pepper_img, (2,0,1))
    x.append(salt_pepper_img)
    y.append(img)

print('x shape',len(x))
print('y shape',len(y))
```

```python
x_train, x_test, y_train, y_test = train_test_split(np.array(x), np.array(y), test_size=0.10, random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
class MyDataset(torch.utils.data.Dataset):
    def __init__(self, x, y):
        self.x = torch.tensor(x)
        self.y = torch.tensor(y)

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        x = self.x[idx]
        y = self.y[idx]
        return (x, y)
    
BATCH_SIZE = 32    
train_dataset = MyDataset(x_train, y_train)
dataLoader_train = torch.utils.data.DataLoader(train_dataset,
                                               batch_size=BATCH_SIZE,
                                               shuffle=True)

test_dataset = MyDataset(x_test, y_test)
dataLoader_test = torch.utils.data.DataLoader(test_dataset,
                                              batch_size=BATCH_SIZE,
                                              shuffle=False)
```

```python
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()

        self.encoder = nn.Sequential(
            nn.Conv2d(3, 32, 4, 1, 2),   # B,  32, 28, 28
            nn.ReLU(True),
            nn.Conv2d(32, 32, 4, 2, 1),  # B,  32, 14, 14
            nn.ReLU(True),
            nn.Conv2d(32, 64, 4, 2, 1),  # B,  64,  7, 7
        )
        
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 4, 2, 1), # B,  64,  14,  14
            nn.ReLU(True),
            nn.ConvTranspose2d(32, 32, 4, 2, 1, 1), # B,  32, 28, 28
            nn.ReLU(True),
            nn.ConvTranspose2d(32, 3, 4, 1, 2)   # B, 1, 28, 28
        )


    def forward(self, x):
        x = self.encoder(x).relu()
        x = self.decoder(x).relu()
        return x
```

```python
model_name = 'salt_pepper_model_mse_0.0001'
model = Autoencoder()

device = 'cuda' if torch.cuda.is_available() else 'cpu'
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.0001)
loss = nn.MSELoss()

trial = Trial(model, optimizer, loss, metrics=['acc', 'loss'],
              callbacks=[callbacks.ConsolePrinter(),
                         make_callback(torchbearer.TARGET, model_name),
                         make_callback(torchbearer.INPUT, model_name),
                         make_callback(torchbearer.PREDICTION, model_name),
                        ]).to(device)

trial.with_generators(train_generator=dataLoader_train, test_generator=dataLoader_test)
history = trial.run(epochs=5, verbose=1)
trial.evaluate(data_key=torchbearer.TEST_DATA)
torch.save(model.state_dict(), './save_{}'.format(model_name))
pd.DataFrame.from_records(history)['loss'].plot()
plt.show()
```
